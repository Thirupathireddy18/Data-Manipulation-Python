# Loading packages 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': {0: 'low', 1: 'high', 2: 'high',3:'low',4:'high'},
                   'B': {0: 'two', 1: 'one', 2: 'two',3:'one',4:'two'},
                   'C': {0: 1, 1: 3, 2: 6, 3: 5, 4: 7},
                   'D': {0: 2, 1: 4, 2: 6,3:6, 4:8},
                   'E':{0: 2,1:3,2:3,3:5,4:6,4:7}})
```

```python
df
```

# Melt function

```python
pd.melt(df, id_vars=['A'], value_vars=['D'])
```

```python
pd.melt(df, id_vars=['A','B'], value_vars=['D','C'])
```

# Pivot Table

```python
table = pd.pivot_table(df, values=['D','C'],index=['A'],columns=['B'], aggfunc=np.sum)
```

```python
table
```

```python
table1 = pd.pivot_table(df, values=['D', 'E'], index=['A','B'],
                    aggfunc={'D': [np.mean,np.median,min,max]
                             ,'E': [np.mean, np.median,min,max]})
```

```python
table1
```
